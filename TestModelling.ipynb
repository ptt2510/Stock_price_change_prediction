```python
from sklearn.preprocessing import MinMaxScaler
fpt_stock_prices = fpt_data.Close.values.astype('float32')
fpt_stock_prices = fpt_stock_prices.reshape(97406,1)
print(fpt_stock_prices.shape)
scaler = MinMaxScaler(feature_range=(0, 1))
fpt_stock_prices = scaler.fit_transform(fpt_stock_prices)
```

```python
train_size = int(len(fpt_stock_prices) * 0.80)
test_size = len(fpt_stock_prices) - train_size
train, test = fpt_stock_prices[0:train_size,:], fpt_stock_prices[train_size:len(fpt_stock_prices),:]
print(len(train), len(test))
```

```python
# Bin ng gi c phiu = Gi c phiu N (pht hoc gi hoc ngy) sau - Gi c phiu hin ti.
# Data Window: make single prediction N step into the future, given N step history
def create_dataset(dataset, N=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-2*N-1):
        a = dataset[i:(i+N), 0]
        dataX.append(a)
        dataY.append(dataset[i+2*N-1, 0])
    return np.array(dataX), np.array(dataY)
```

```python
import numpy as np
# Bi ton: input: Cung cp gi c phiu hin ti v N-1 gi c phiu trc , u ra s l d on Gi c phiu N (pht hoc gi hoc ngy) sau
# chn N=5
N = 5
trainX, trainY = create_dataset(train, N)
testX, testY = create_dataset(test, N)
```

```python
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1],1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1],1))
```

```python
#build model lstm
from keras.layers import Dense, Activation, Dropout
from keras.layers import LSTM
from keras.models import Sequential
import time
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape= (trainX.shape[1], 1)))
model.add(LSTM(64, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

start = time.time()
model.compile(loss='mse', optimizer='adam')
print ('compilation time : ', time.time() - start)
```

```python
model.fit(
    trainX,
    trainY,
    batch_size=128,
    epochs=10,
    validation_split=0.05)
```

```python
y_pred=model.predict(testX)
mean_squared_error(testY, y_pred)
```

```python
model.save('LSTM.h5')
print('Model Saved!')
```

```python
# load model
from tensorflow.keras.models import load_model
savedModel=load_model('LSTM.h5')
savedModel.summary()
```

```python
from numpy import newaxis
def predict_stock_change_after_N_minutesORhoursORdays(model, inp):
    stock_price_predict = model.predict(inp[newaxis,:,:])
    stock_price_predict= scaler.inverse_transform(stock_price_predict.reshape(-1, 1))
    current_value=inp[4]
    current_value=scaler.inverse_transform(current_value.reshape(-1, 1))
    stock_price_change= float(stock_price_predict) - float(current_value)
    return stock_price_change
 
```

```python
stock_price_change = predict_stock_change_after_N_minutesORhoursORdays(model, testX[0])
print('predict stock change after N minutesORhoursORdays: '+ str(stock_price_change))
```

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly_train = poly_features.fit_transform(trainX.reshape(trainX.shape[0],trainX.shape[1]))
X_poly_test = poly_features.fit_transform(testX.reshape(testX.shape[0],testX.shape[1]))
```

```python
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(warm_start=True, penalty=None,learning_rate="constant", eta0=0.0005)
#early stopping
from sklearn.base import clone
minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_poly_train, trainY)  # continues where it left off
    y_val_predict = sgd_reg.predict(X_poly_test)
    val_error = mean_squared_error(y_val_predict, testY)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
print(best_epoch)
print(minimum_val_error)
```

```python
# Polynomial Regression
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly_train = poly_features.fit_transform(trainX.reshape(trainX.shape[0],trainX.shape[1]))
X_poly_test = poly_features.fit_transform(testX.reshape(testX.shape[0],testX.shape[1]))
from sklearn.linear_model import SGDRegressor
sgd_reg = SGDRegressor(warm_start=True, penalty=None,learning_rate="constant", eta0=0.0005)
#early stopping
from sklearn.base import clone
minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_poly_train, trainY)  # continues where it left off
    y_val_predict = sgd_reg.predict(X_poly_test)
    val_error = mean_squared_error(y_val_predict, testY)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
print(best_epoch)
print(minimum_val_error)
```
